```python
import pandas as pd
from tqdm import tqdm
import ktest
from ktest.tester import Ktest
```

# Ktest: two-sample kernel tests for large datasets
## Univariate testing: example

Loading data:

```python
data = pd.read_csv('v5_data/RTqPCR_reversion_logcentered.csv', index_col=0)
meta = pd.Series(data = pd.Series(data.index).apply(lambda x : x.split(sep='.')[1]))
meta.index = data.index
```

Performing ktest in the univariate setting on every gene:

```python
univ_test_stats = pd.DataFrame()
univ_test_pvalues = pd.DataFrame()
it = tqdm(range(len(data.columns)))
for i in it:
    kt = Ktest(data=data.iloc[:, i], metadata=meta, sample_names=['48HREV','48HDIFF'])
    kt.test(verbose=0)
    univ_test_stats[data.columns[i]] = kt.kfda_statistic
    univ_test_pvalues[data.columns[i]] = kt.kfda_pval_asymp
#univ_test_stats.index + 1
#univ_test_pvalues.index + 1
```

Calculate the Benjamini-Hochberg adjusted p-value:

```python
def correct_BenjaminiHochberg_pval_of_dfcolumn(df):
    df = pd.concat([df,df.rank()],keys=['pval','rank'],axis=1)
    df['pvalc'] = df.apply(lambda x: len(df) * x['pval']/x['rank'],axis=1) # correction
    df['rankc'] = df['pvalc'].rank() # new order calculation
    corrected_pvals = []
    # Correction of p-values that changed order:
    l = []
    if not df['rankc'].equals(df['rank']):
        first_rank = df['rank'].sort_values().values[0]
        pvalc_prec = df.loc[df['rank']==first_rank,'pvalc'].iat[0]
        df['rank'] = df['rank'].fillna(10000)
        for rank in df['rank'].sort_values().values[1:]:
            pvalc = df.loc[df['rank']==rank,'pvalc'].iat[0]
            if pvalc_prec >= 1 : 
                pvalue = 1 # l += [1]
            elif pvalc_prec > pvalc :
                pvalue = pvalc
            elif pvalc_prec <= pvalc:
                pvalue = pvalc_prec
            else: 
                print('error pval correction',f'rank{rank} pvalc{pvalc} pvalcprec{pvalc_prec}')
                print(df.loc[df['rank']==rank].index)
            pvalc_prec = pvalc
            l += [pvalue] 
        pvalue = 1 if pvalc >1 else pvalc
        l += [pvalue]
    if len(l)>0: 
        return(pd.Series(l,index=df['rank'].sort_values().index))
    else: 
        return(pd.Series(df['pvalc'].values,index=df['rank'].sort_values().index))
```

```python
t = 10
t_pvalues_corrected = correct_BenjaminiHochberg_pval_of_dfcolumn(univ_test_pvalues.loc[t])
print("Top 5 DE genes (with BH correction): \n")
for g in t_pvalues_corrected.index[:5]:
    print(g)
```
